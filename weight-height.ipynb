```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data=pd.read_csv('weight-height.csv')
data.head()
```

```python
data.shape
```

```python
data.plot(kind='scatter',x='Height',y='Weight')
plt.show()
```

```python
data.plot(kind='box')
```

```python
data.corr()
```

```python
weight=pd.DataFrame(data['Weight'])
height=pd.DataFrame(data['Height'])
weight
```

```python
height
```

```python
data.shape
```

```python
lm=LinearRegression()
model=lm.fit(weight,height)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(weight,height)
```

```python
x=([67,87,92])
x=pd.DataFrame(x)
y=model.predict(x)
y=pd.DataFrame(y)
df=pd.concat([x,y],axis=1,keys=['weight_new','height_predicted'])
df
```

```python
data.plot(kind='scatter',x='Height',y='Weight')
plt.plot(weight,model.predict(weight),color='red',linewidth=4)
plt.scatter(x,y)
plt.show()
```
